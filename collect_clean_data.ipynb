# 2. Clean dataset:

## a. Weather data from airport:

```python
df_weather = pd.read_csv("./data/weather_laguardia_airport_2024.csv")
df_weather.head(5)
```

```python
df_weather.columns
```

```python
weather_necessary_columns = [
    "DATE",
    "WND",
    "VIS",
    "TMP",
    "REM",
]
df_weather = df_weather[weather_necessary_columns]
df_weather.head()
```

### Missing values:

```python
df_weather.isnull().sum().sort_values(ascending=False)
```

### Duplicated values: 

```python
df_weather.duplicated().sum()
```

### Convert types:

```python
df_weather.info()
```

```python
df_weather.iloc[0:1]
```

```python
# Convert 'Date' column to datetime format
df_weather['DATE'] = pd.to_datetime(df_weather['DATE'])
# Convert WND, VIS, TEMP to appropriate numeric types:
df_weather['WND'] = df_weather['WND'].str.split(',').str[3].astype(float) / 10
df_weather['visibility_km'] = df_weather['VIS'].str.split(',').str[0].astype(float) / 1000
df_weather['temperature_C'] = df_weather['TMP'].str.split(',').str[0].str.replace('+', '', regex=False).astype(float) / 10
```
