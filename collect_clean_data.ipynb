# 1. Concatenate the data from multiple Parquet files into a single DataFrame:

```python
import glob
import pandas as pd

file_list = glob.glob("./data/yellow_tripdata_2024-*.parquet")
file_list

df_trip = pd.concat(
    [pd.read_parquet(file) for file in file_list],
    ignore_index=True,
    sort=False
)

# Read it into the DataFrame
df_trip.to_csv("./data/yellow_tripdata_2024.csv", index=False)

# Display the first 5 rows of the DataFrame
df_trip.head(5)
```
