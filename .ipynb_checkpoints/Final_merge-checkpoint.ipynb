## Final Merge of Trip and Weather Datasets

```python
import pandas as pd

# Load cleaned trip data
trip_agg = pd.read_csv("./data/cleaned_yellow_trip_hourly.csv")
trip_agg['date'] = pd.to_datetime(trip_agg['date']).dt.date
trip_agg.head()
```

```python
#Load weather data
weather_clean = pd.read_csv("./data/cleaned_weather_hourly.csv")
weather_clean['date'] = pd.to_datetime(weather_clean['date']).dt.date

weather_clean.head()
```

```python
# Merge with trip data
merged_df = pd.merge(trip_agg, weather_clean, on=['date', 'hour'], how='inner')

# Remove duplicate rows (if any)
merged_df = merged_df.drop_duplicates()

# Save merged CSV
merged_df.to_csv('./data/final_taxi_weather.csv', index=False)


# Display the first 5 rows of the DataFrame
merged_df.head()
```
